```python
import pandas as pd
import numpy as np
import random
import os
```

```python
random.seed(42)
np.random.seed(42)
# Show all columns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
```

```python
data_dir = r'data\ssusa'
os.makedirs(data_dir, exist_ok=True)

save_dir = data_dir + r'\species_list_cluster'
os.makedirs(save_dir, exist_ok=True)

df_path = data_dir + r'\Data_Clustered_Cleaned_50km_WithSciName.csv'
df = pd.read_csv(df_path)

df.head()
```

```python
# Count all unique species in the dataframe
unique_species_list = df['Sci_Name'].dropna().unique()
unique_species_list = sorted(unique_species_list)
unique_species_count = len(unique_species_list)
print(f"Total unique species: {unique_species_count}")
```

```python

# 1. Group by Cluster only, and compile a list of unique species (Sci_Name)
species_list_by_cluster = (
    df.groupby(['Cluster_Agglo'])['Sci_Name']
    .unique() 
    .reset_index()
    .rename(columns={'Sci_Name': 'Species_List'})
)


# 2. Group by Cluster and year, and compile a list of unique species (Sci_Name)
species_list_by_cluster_year = (
    df.groupby(['Cluster_Agglo', 'Year'])['Sci_Name']
    .unique() 
    .reset_index()
    .rename(columns={'Sci_Name': 'Species_List'})
)
```

```python
species_list_by_cluster.head()
```

```python
species_list_by_cluster_year.head()
```

```python
# Get cluster info
cluster_info = df[['Year', 'Cluster_Agglo', 'Centroid_Latitude', 'Centroid_Longitude', 'City_State']].drop_duplicates()
cluster_info_no_year = df[['Cluster_Agglo', 'Centroid_Latitude', 'Centroid_Longitude', 'City_State']].drop_duplicates()

# Merge on cluster only
df_species_list = pd.merge(
    species_list_by_cluster,
    cluster_info_no_year,
    on=['Cluster_Agglo'],
    how='left',
    suffixes=('', '_cluster')
)
df_species_list['Richness'] = df_species_list['Species_List'].apply(len)


# Merge on cluster and Year
df_species_list_year = pd.merge(
    species_list_by_cluster_year,
    cluster_info,
    on=['Cluster_Agglo', 'Year'],
    how='left',
    suffixes=('', '_cluster')
)
df_species_list_year['Richness'] = df_species_list_year['Species_List'].apply(len)
```

```python
df_species_list.head()
```

```python
df_species_list_year.head()
```

```python
# Build a DataFrame for one-hot encoding
one_hot = pd.DataFrame(
    [
        [int(species in species_list) for species in unique_species_list]
        for species_list in df_species_list['Species_List']
    ],
    columns=unique_species_list,
    index=df_species_list.index
)

one_hot_df = pd.concat([df_species_list[['Cluster_Agglo', 'Centroid_Latitude', 'Centroid_Longitude', 'City_State', 'Richness']], one_hot], axis=1)
```

```python
one_hot_df.head()
```

```python
# Build a DataFrame for one-hot encoding
one_hot_year = pd.DataFrame(
    [
        [int(species in species_list) for species in unique_species_list]
        for species_list in df_species_list_year['Species_List']
    ],
    columns=unique_species_list,
    index=df_species_list_year.index
)

one_hot_year_df = pd.concat([df_species_list_year[['Cluster_Agglo', 'Year', 'Centroid_Latitude', 'Centroid_Longitude', 'City_State', 'Richness']], one_hot_year], axis=1)
```

```python
one_hot_year_df.head()
```

```python
# Convert the 'Species_List' column in df_species_list to a comma-separated string for each row
df_species_list['Species_List'] = df_species_list['Species_List'].apply(lambda x: ', '.join(x))
df_species_list_year['Species_List'] = df_species_list_year['Species_List'].apply(lambda x: ', '.join(x))
```

```python
# Add a 'Radius' column after 'Centroid_Latitude' and 'Centroid_Longitude' for all relevant DataFrames

# Define a default radius value (change as needed)
default_radius = 50

# For df_species_list
if 'Radius_Km' not in df_species_list.columns:
    insert_at = df_species_list.columns.get_loc('Centroid_Longitude') + 1
    df_species_list.insert(insert_at, 'Radius_Km', default_radius)

# For df_species_list_year
if 'Radius_Km' not in df_species_list_year.columns:
    insert_at = df_species_list_year.columns.get_loc('Centroid_Longitude') + 1
    df_species_list_year.insert(insert_at, 'Radius_Km', default_radius)

# For one_hot_df
if 'Radius_Km' not in one_hot_df.columns:
    insert_at = one_hot_df.columns.get_loc('Centroid_Longitude') + 1
    one_hot_df.insert(insert_at, 'Radius_Km', default_radius)

# For one_hot_year_df
if 'Radius_Km' not in one_hot_year_df.columns:
    insert_at = one_hot_year_df.columns.get_loc('Centroid_Longitude') + 1
    one_hot_year_df.insert(insert_at, 'Radius_Km', default_radius)
```

```python
# Add Habitat info to the relevant DataFrames, inserting after City_State

# For df_species_list
if 'Habitat' not in df_species_list.columns:
    # Get all unique habitats for each cluster, join as string
    habitat_map = df.groupby('Cluster_Agglo')['Habitat'].agg(lambda x: ', '.join(sorted(set(x.dropna()))))
    insert_at = df_species_list.columns.get_loc('City_State') + 1
    df_species_list.insert(insert_at, 'Habitat', df_species_list['Cluster_Agglo'].map(habitat_map))

# For df_species_list_year
if 'Habitat' not in df_species_list_year.columns:
    # Get all unique habitats for each (cluster, year) pair, join as string
    habitat_map_year = df.groupby(['Cluster_Agglo', 'Year'])['Habitat'].agg(lambda x: ', '.join(sorted(set(x.dropna()))))
    insert_at = df_species_list_year.columns.get_loc('City_State') + 1
    df_species_list_year.insert(
        insert_at,
        'Habitat',
        df_species_list_year.set_index(['Cluster_Agglo', 'Year']).index.map(habitat_map_year)
    )

# For one_hot_df
if 'Habitat' not in one_hot_df.columns:
    insert_at = one_hot_df.columns.get_loc('City_State') + 1
    one_hot_df.insert(insert_at, 'Habitat', one_hot_df['Cluster_Agglo'].map(habitat_map))

# For one_hot_year_df
if 'Habitat' not in one_hot_year_df.columns:
    insert_at = one_hot_year_df.columns.get_loc('City_State') + 1
    one_hot_year_df.insert(
        insert_at,
        'Habitat',
        one_hot_year_df.set_index(['Cluster_Agglo', 'Year']).index.map(habitat_map_year)
    )
```

```python
# output the one hot encoded dataframe to csvs
df_species_list.to_csv(save_dir + r'\species_list_cluster_df.csv', index=False)
df_species_list_year.to_csv(save_dir + r'\species_list_cluster_year_df.csv', index=False)
one_hot_df.to_csv(save_dir + r'\species_list_cluster_one_hot.csv', index=False)
one_hot_year_df.to_csv(save_dir + r'\species_list_cluster_year_one_hot.csv', index=False)
```

```python
# Create a DataFrame where each unique species is a row, and add a column with the list of clusters in which the species occurs

# For each species, find the unique clusters it appears in
species_cluster_map = {
    species: sorted(df.loc[df['Sci_Name'] == species, 'Cluster_Agglo'].unique().tolist())
    for species in unique_species_list
}

# Build the DataFrame
species_cluster_df = pd.DataFrame({
    'Sci_Name': list(species_cluster_map.keys()),
    'Cluster_List': list(species_cluster_map.values())
})

```

```python
# Add Class, Order, Family, Genus, Species, Common_Name info to species_cluster_df

# Get unique species info from the original dataframe
species_info_cols = ['Sci_Name', 'Class', 'Order', 'Family', 'Genus', 'Species', 'Common_Name']
species_info_df = df[species_info_cols].drop_duplicates(subset=['Sci_Name'])

# Merge species_cluster_df with species_info_df on 'Sci_Name'
species_cluster_df = species_cluster_df.merge(species_info_df, on='Sci_Name', how='left')

# Reorder columns for clarity
cols = ['Sci_Name', 'Class', 'Order', 'Family', 'Genus', 'Species', 'Common_Name', 'Cluster_List']
species_cluster_df = species_cluster_df[cols]
```

```python
species_cluster_df.head()
```

```python
# Explode the Cluster_List so each row is a species-year-cluster
species_cluster_df_exploded = species_cluster_df.explode('Cluster_List')

# Create one-hot encoding for clusters
one_hot = pd.get_dummies(species_cluster_df_exploded['Cluster_List'], prefix='Cluster')

# Combine with original data
species_cluster_df_one_hot = pd.concat([species_cluster_df_exploded, one_hot], axis=1)

# Group by original index (or by Sci_Name, Year, etc.) and sum to collapse back to one row per species-year
species_cluster_df_one_hot_final = species_cluster_df_one_hot.groupby(['Sci_Name', 'Class', 'Order', 'Family', 'Genus', 'Species', 'Common_Name'], as_index=False)[one_hot.columns].sum()
```

```python
species_cluster_df_one_hot_final.head()
```

```python
# Create a DataFrame where each row is a unique (Sci_Name, Year) pair, with the list of clusters in which the species occurs that year

# For each (species, year), find the unique clusters it appears in
species_year_cluster_rows = []
for species in unique_species_list:
    subdf = df[df['Sci_Name'] == species]
    for year in sorted(subdf['Year'].unique()):
        clusters = sorted(subdf.loc[subdf['Year'] == year, 'Cluster_Agglo'].unique().tolist())
        species_year_cluster_rows.append({
            'Sci_Name': species,
            'Year': year,
            'Cluster_List': clusters
        })

species_year_cluster_df = pd.DataFrame(species_year_cluster_rows)

# Merge with species info
species_year_cluster_df = species_year_cluster_df.merge(species_info_df, on='Sci_Name', how='left')

# Reorder columns for clarity
cols = ['Sci_Name', 'Year', 'Class', 'Order', 'Family', 'Genus', 'Species', 'Common_Name', 'Cluster_List']
species_year_cluster_df = species_year_cluster_df[cols]
```

```python
species_year_cluster_df.head()
```

```python
# Explode the Cluster_List so each row is a species-year-cluster
species_year_cluster_df_exploded = species_year_cluster_df.explode('Cluster_List')

# Create one-hot encoding for clusters
one_hot = pd.get_dummies(species_year_cluster_df_exploded['Cluster_List'], prefix='Cluster')

# Combine with original data
species_year_cluster_df_one_hot = pd.concat([species_year_cluster_df_exploded, one_hot], axis=1)

# Group by original index (or by Sci_Name, Year, etc.) and sum to collapse back to one row per species-year
species_year_cluster_df_one_hot_final = species_year_cluster_df_one_hot.groupby(['Sci_Name', 'Year', 'Class', 'Order', 'Family', 'Genus', 'Species', 'Common_Name'], as_index=False)[one_hot.columns].sum()
```

```python
species_year_cluster_df_one_hot_final.head()
```

```python
# Convert Cluster_List to string for CSV output for both dataframes: species_year_cluster_df and species_cluster_df
species_year_cluster_df['Cluster_List'] = species_year_cluster_df['Cluster_List'].apply(lambda x: ','.join(map(str, x)) if isinstance(x, list) else str(x))
species_cluster_df['Cluster_List'] = species_cluster_df['Cluster_List'].apply(lambda x: ','.join(map(str, x)) if isinstance(x, list) else str(x))
```

```python

# Output the dataframes to CSV
species_year_cluster_df.to_csv(f"{save_dir}/species_year_cluster_table.csv", index=False)
species_cluster_df.to_csv(f"{save_dir}/species_cluster_table.csv", index=False)
species_cluster_df_one_hot_final.to_csv(f"{save_dir}/species_cluster_onehot_table.csv", index=False)
species_year_cluster_df_one_hot_final.to_csv(f"{save_dir}/species_year_cluster_onehot_table.csv", index=False)
```
