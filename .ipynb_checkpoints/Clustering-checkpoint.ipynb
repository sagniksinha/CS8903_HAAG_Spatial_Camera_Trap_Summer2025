```python
import numpy as np
import matplotlib.pyplot as plt
from kneed import KneeLocator
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import gc

# Step 1: UTM projection (example uses UTM zone 33N, modify if needed)
transformer = Transformer.from_crs("epsg:4326", "epsg:32633", always_xy=True)
merged_df['x'], merged_df['y'] = transformer.transform(
    merged_df['Longitude'].values,
    merged_df['Latitude'].values
)

# Step 2: Prepare data
coords = merged_df[['x', 'y']].values

# Step 3: Run KMeans for k=2 to 20 and store inertia and silhouette scores
inertia = []
sil_scores = []
k_range = range(2, 21)

for k in k_range:
    print(f"Running KMeans for k={k}")
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(coords)
    inertia.append(kmeans.inertia_)
    labels = kmeans.labels_
    score = silhouette_score(coords, labels, sample_size=min(25000), random_state=0)
    sil_scores.append(score)

# Step 4: Detect elbow with KneeLocator
knee = KneeLocator(k_range, inertia, curve='convex', direction='decreasing', S=1.0)
print(f"Optimal k (elbow) detected at: {knee.elbow}")

# Optional: plot normalized knee curve
knee.plot_knee_normalized()
plt.show()

# Step 5: Plot elbow chart with vertical line at elbow
plt.figure(figsize=(8, 5))
plt.plot(k_range, inertia, 'bo-', label='Inertia')
if knee.elbow is not None:
    plt.axvline(x=knee.elbow, color='red', linestyle='--', label=f'Elbow at k={knee.elbow}')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia (Sum of squared distances)')
plt.title('Elbow Method for Optimal k')
plt.grid(True)
plt.legend()
plt.show()

# Step 6: Plot % reduction in inertia (first derivative)
inertia_diff = np.diff(inertia)
inertia_pct_drop = 100 * (-inertia_diff / inertia[:-1])
plt.figure(figsize=(8, 5))
plt.plot(k_range[1:], inertia_pct_drop, 'go-', label='Inertia % drop')
plt.xlabel('k')
plt.ylabel('% Drop in Inertia')
plt.title('Rate of Change in Inertia')
plt.grid(True)
plt.legend()
plt.show()

# Step 7: Silhouette score plot
plt.figure(figsize=(8, 5))
plt.plot(k_range, sil_scores, 'mo-', label='Silhouette Score')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Method for Optimal k')
plt.grid(True)
plt.legend()
plt.show()

gc.collect()
```

```python
from pyproj import Transformer
from sklearn.cluster import KMeans

# Step 1: UTM projection (example uses UTM zone 33N, modify if needed)
transformer = Transformer.from_crs("epsg:4326", "epsg:32633", always_xy=True)
merged_df['x'], merged_df['y'] = transformer.transform(
    merged_df['Longitude'].values,
    merged_df['Latitude'].values
)

# Step 2: Run KMeans with k=8 on projected coordinates
k = 20
kmeans = KMeans(n_clusters=k, random_state=42)
merged_df['cluster_label'] = kmeans.fit_predict(merged_df[['x', 'y']].values)

# Step 3: Save result to CSV
# Define file path
output_path = save_dir + r'\merged_snapshot_usa_with_label.csv'
merged_df.to_csv(output_path, index=False)

print(f"KMeans clustering done with k={k}. Saved to {output_path}")
```

```python
# Create a GeoDataFrame from your dataframe with lat/lon points
geometry = [Point(xy) for xy in zip(merged_df['Longitude'], merged_df['Latitude'])]
gdf = gpd.GeoDataFrame(merged_df, geometry=geometry, crs="EPSG:4326")

# Convert to Web Mercator for plotting with contextily basemap
gdf = gdf.to_crs(epsg=3857)

# Plot
fig, ax = plt.subplots(figsize=(12, 10))

# Scatter plot colored by cluster label
gdf.plot(column='cluster_label', categorical=True, cmap='tab20', legend=True, markersize=10, alpha=0.7, ax=ax)

# Add basemap (OpenStreetMap)
ctx.add_basemap(ax, source=ctx.providers.CartoDB.Positron)

# Set title and axes off
ax.set_title('KMeans Clusters on North America Map', fontsize=16)
ax.set_axis_off()

plt.show()
```
